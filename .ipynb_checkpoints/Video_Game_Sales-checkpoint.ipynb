```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()
```

```python
game_sales = pd.read_csv('top100videogamesales.csv')
```

----

```python
sales_by_publisher = game_sales.groupby('Publisher').sum()
region_sales_by_publisher = sales_by_publisher[['NA_Sales', 'EU_Sales','JP_Sales','Other_Sales']]
region_sales_by_publisher.plot(kind='bar', stacked=True, figsize=(9,4))
plt.xlabel('Publisher')
plt.title('Regional Video Game Sales by Publisher', fontsize=20)
plt.ylabel('Sales (millions)')
plt.xticks(rotation=70)
plt.legend(title='Region')
plt.show()
```

The chart visualizes the regional distribution of video game sales for various publishers. Each bar represents a publisher, and the segments within each bar indicate the sales in different regions: North America (NA), Europe (EU), Japan (JP), and Other regions.

## Analysis
- **Nintendo** stands out as the publisher with the highest total sales. The vast majority of their sales come from North America and Europe, which form the bulk of the stacked bar for Nintendo.
- **Activision** follows but with a significantly smaller total sales volume, showing a similar distribution pattern to Nintendo, with North America and Europe being the largest markets.
- **Take-Two Interactive** and other publishers like **Electronic Arts** and **Microsoft Game Studios** also show notable sales figures, primarily driven by the North American and European markets.

## Conclusion
- The dominance of **Nintendo** in the global video game market is evident from the chart, with their total sales far surpassing other publishers.
- The North American and European markets are the largest contributors to sales for these top publishers, which is consistent across the chart.

## Publisher with the Most Sales
- **Nintendo** has the most sales, with a total of **653.18 million units** sold globally. This is a significant lead over the next closest publisher, Activision, which sold **179.90 million units** globally.

```python
global_sales_per_year = game_sales.groupby('Year')['Global_Sales'].sum()
global_sales_per_year.plot(kind='line', marker='o', figsize=(9,4))
plt.title('Global Game Sales By Year', fontsize=20)
plt.xlabel('Year')
plt.ylabel('Total Sales (Millions)')
plt.grid(True)
plt.xticks(rotation=70)
plt.show()
```

# Global Video Game Sales by Year

The line chart represents the total global sales of video games for each year in the dataset. The x-axis shows the years, and the y-axis represents the total number of units sold globally, in millions.

## Observations
- Peaks in the chart may indicate years with particularly high sales, which could be due to various factors like the release of major gaming titles or gaming consoles.
- Troughs or declines might suggest years with fewer hit titles or changes in market conditions.
- Any upward or downward trends could indicate the overall health and growth of the video game industry over time.

```python
action_games = game_sales[game_sales['Genre'] == 'Action']
plt.figure(figsize=(10, 6))
plt.hist(action_games['Global_Sales'], bins=22, color='blue', edgecolor='black')
plt.title('Global Sales Distribution for Action Games', fontsize=20)
plt.xlabel('Global Sales (Millions)')
plt.ylabel('Number of Games')
plt.grid(True)
plt.show()
```

# Global Sales Distribution for Action Games

The histogram displays the global sales distribution for video games in the Action genre, with sales ranges on the x-axis and the count of games on the y-axis.

## Observations from the Histogram
- The majority of Action games achieve sales between 6 to 12 million units, with a noticeable peak in the 6 to 8 million range.
- There is a significant drop in the number of games with sales between 8 to 10 and 12 to 14 million units, indicating these sales figures are less common for Action games.
- No Action games fall within the 14 to 16 and 18 to 20 million units sales ranges, as shown by the gaps in the histogram.
- A few Action games show high sales figures, with 2 games in both the 16 to 18 and 20 to 22 million units ranges, suggesting that while rare, there are Action titles with exceptional sales performance.

```python
na_sales_by_genre = game_sales.groupby('Genre')['NA_Sales'].sum()
threshold = na_sales_by_genre.quantile(0.3)
small_genres = na_sales_by_genre[na_sales_by_genre < threshold]
na_sales_by_genre = na_sales_by_genre[na_sales_by_genre >= threshold]
na_sales_by_genre['Other'] = small_genres.sum()
plt.figure(figsize=(10, 8))
plt.pie(na_sales_by_genre, labels=na_sales_by_genre.index, autopct='%1.1f%%', startangle=140)
plt.title('North American Video Game Sales by Genre', fontsize=20)
plt.show()
```

## North American Video Game Sales by Genre

The pie chart illustrates the distribution of video game sales in North America by genre. Notable insights include:

- **Shooter**: The most significant share at 24.8%, indicating a strong preference for this genre.
- **Sports**: At 15.0%, this genre also shows substantial popularity.
- **Action**: Holding 13.8%, demonstrating its steady appeal.

Genres such as **Misc** (12.5%), **Platform** (9.3%), **Racing** (10.0%), and **Role-Playing** (8.9%) have noteworthy portions of the market.

### The "Other" Category
This category comprises 5.7% of the total and includes genres that did not individually meet the 3% threshold. These are:

- **Fighting**
- **Simulation**
- **Puzzle**


---------------------

```python
eu_sales_by_genre = game_sales.groupby('Genre')['EU_Sales'].sum()
threshold_eu = eu_sales_by_genre.quantile(0.3)
small_genres_eu = eu_sales_by_genre[eu_sales_by_genre < threshold_eu]
eu_sales_by_genre = eu_sales_by_genre[eu_sales_by_genre >= threshold_eu]
eu_sales_by_genre['Other'] = small_genres_eu.sum()
plt.figure(figsize=(10, 8))
plt.pie(eu_sales_by_genre, labels=eu_sales_by_genre.index, autopct='%1.1f%%', startangle=140)
plt.title('European Video Game Sales by Genre', fontsize=20)
plt.show()
```

## European Video Game Sales by Genre

The pie chart illustrates the distribution of video game sales across various genres within the European market. The approach taken simplifies the visualization by aggregating less prevalent genres into an "Other" category. Notable observations are:

- **Shooter** and **Sports** genres lead the market, each occupying a substantial segment, indicative of their popularity.
- **Action** games also hold a considerable portion, reflecting their wide appeal.
- Other prominent genres include **Misc**, **Platform**, **Racing**, and **Role-Playing**, each maintaining a significant presence.

### "Other" Category Composition
This category, representing genres with less than 3% of sales, includes:

- **Fighting**
- **Puzzle**
- **Simulation**



----------------------------------------------------------

## Comparison of Video Game Sales by Genre: North America vs Europe


### Similarities:
1. **Dominant Genres**: In both markets, the **Shooter**, **Sports**, and **Action** genres command significant market shares. This similarity indicates a cross-regional appeal for these types of games.
2. **Diversity in Preferences**: Both North American and European markets show diversity in genre preferences, with **Misc**, **Platform**, **Racing**, and **Role-Playing** games enjoying considerable popularity in both regions.

### Differences:
1. **Leading Genre**: In North America, the **Shooter** genre has the largest share (24.8%), while in Europe, both **Shooter** and **Sports** genres are almost equally dominant. This suggests a slightly varied preference between the two regions.
2. **Proportional Differences**: Although the same genres are popular in both regions, their proportions vary. For example, **Shooter** games hold a notably larger percentage in North America compared to Europe.

```python
games_by_platform = game_sales['Platform'].value_counts()
sorted_games_by_platform = games_by_platform.sort_values(ascending=False)
plt.figure(figsize=(12, 8))
sorted_games_by_platform.plot(kind='bar', color='skyblue')
plt.title('Count of Games by Platform', fontsize=20)
plt.xlabel('Platform', fontsize=15)
plt.ylabel('Number of Games', fontsize=15)
plt.xticks(rotation=45)
plt.show()
```

## Count of Games by Platform

The bar chart provides an overview of the total number of games created for each platform within the dataset. Key observations include:

- **Nintendo**: Emerges as the leading platform with the highest number of games, indicating its significant role in the gaming industry.
- **PlayStation**: Follows Nintendo, showcasing its strong presence and popularity among game developers and players.
- **XBox**: Holds a considerable number of games, though less than Nintendo and PlayStation, reflecting its market position.
- **Other Platforms**: There's a minimal representation of games from other platforms, suggesting a concentrated market around the three major platforms.

### Conclusion
The chart highlights the dominance of Nintendo, PlayStation, and XBox in the gaming market, with Nintendo leading in terms of the number of games available. This distribution reflects the market trends and preferences in the gaming industry.
